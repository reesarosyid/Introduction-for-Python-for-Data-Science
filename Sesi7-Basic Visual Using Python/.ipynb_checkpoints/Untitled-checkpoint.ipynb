# Basic Visualization using Python

## Import library

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use(['ggplot'])
```

## Import dataset

```python
df = pd.read_excel("Canada.xlsx",
                   sheet_name='Canada by Citizenship',
                   skiprows=20,
                   skipfooter=2)
```

## Menampilkan data

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

## Data Cleansing

Menghapus kolom yang tidak diperlukan seperti AREA, REG, DEV, Type, Coverage.

```python
kol = ['AREA', 'REG', 'DEV', 'Type', 'Coverage']
df.drop(kol, axis=1, inplace=True)
df.head()
```

Merubah nama OdName, AreaName, RegName menjadi Country, Continent, Region

```python
df.rename(columns={'OdName':'Country',
                       'AreaName':'Continent',
                       'RegName':'Region'}, inplace=True)
```

```python
df.head()
```

Membuat kolom total

```python
df['Total'] = df.sum(axis=1)
```

```python
df.tail()
```

Membuat value country sebagai index

```python
df.set_index('Country', inplace=True)
df.head()
```

## Plotting

### Line Plot

Line plot adalah salah satu jenis visualisasi data yang banyak digunakan dan merupakan jenis plot dasar dalam visualisasi data. Jenis plot ini menampilkan informasi berupa rangkaian titik data yang terhubung dengan segmen garis lurus.

**Pertanyaan** : Data imigran Haiti dari tahun 1980 - 2014

```python
years = list(range(1980,2014))
haiti = df.loc['Haiti', years]
haiti.head()
```

```python
haiti.plot()

plt.title("Imigran from Haiti")
plt.ylabel("Total Imigran")
plt.xlabel("Years")

#Memberi keterangan teks pada plot
plt.text(1995, 6000, '2010 Haiti Earthquake')
```

**Pertanyaan** : Bagaimana perbandingan jumlah imigran China dan India dari tahun 1980 - 2013 ke Canada?

```python
cntry = ['China', 'India']
china_india = df.loc[cntry, years]
china_india
```

```python
china_india.plot()
```

Plotnya masih salah karena kolom country masih menjadi baris belum menjadi kolom. Agar benar maka harus ditranspose.

```python
CI_T = china_india.T
CI_T.head()
```

```python
CI_T.plot()
plt.title("Imigran form China and India")
plt.ylabel("Total Imigran")
plt.xlabel("Years")
```

**Pertanyaan 2** : Bandingkan tren dari 5 negara teratas yang menyumbang imigran ke Kanada

```python
top5 = df['Total'].sort_values(ascending=False).head().index
top5 = df.loc[top5,years]
top5 = top5.T
```

```python
top5.plot(figsize=(15,8))
plt.title("Immigration from top 5 countries")
plt.ylabel("Total Imigran")
plt.xlabel("Years")
```

### Area Plot 

Diagram area adalah variasi dari diagram garis atau diagram deret waktu. Diagram ini menggunakan daerah berbayang di bawah garis yang diplot pada grafik untuk menunjukkan volume data yang diwakili oleh garis. Seperti diagram rangkaian waktu, diagram area berguna untuk melihat trend data dari waktu ke waktu.

```python
top5.plot(kind = 'area',
          stacked = False,
           figsize = (15,8))
plt.title("Immigration from top 5 countries")
plt.ylabel("Total Imigran")
plt.xlabel("Years")
```

### Histogram

Histogram adalah representasi grafis untuk distribusi warna dari citra digital. Sumbu ordinat vertikal merupakan representasi piksel dengan nilai tonal dari tiap-tiap deret bin pada sumbu axis horizontalnya.

**Pertanyaan** : Bagaimana distribusi dari Imigran Denmark, Norwegia dan Swedia ?

```python
df_dns = df.loc[["Denmark", "Norway", "Sweden"], years]
df_dns
```

```python
df_dns.T.plot(kind='hist',
              alpha=0.5,
              bins=15,
              figsize=(10,6),
              color=['coral','darkslateblue','mediumseagreen'])
```

### Barplot

Bar graph atau bar chart merupakan grafik yang berfungsi menunjukkan frekuensi atau fraksi dari data kategori. Selain frekuensi atau fraksi, bar plot juga dapat digunakan untuk menjelaskan nilai kuantitatif lain untuk menjelaskan kategori pada data seperti mean dan persen perubahan.

```python
top5.sum()
```

```python
top5.sum().plot(kind='bar')
```

```python
top5.sum().plot(kind='barh');
```

### Pie Chart

Diagram pie dapat digunakan untuk perbandingan komponen. Perbandingan komponen menunjukkan ukuran setiap komponen sebagai persentase dari total.

**Pertanyaan** : Perbandingan Benua

```python
df_continents = df.groupby('Continent').sum()
df_continents
```

```python
df_continents['Total'].plot(kind='pie',
                            figsize=(5,6),
                            autopct="%.2f%%",
                            startangle=90,
                            shadow=True)
```

```python
explode_list = [0.1,0,0,0,0.1,0.1]
df_continents['Total'].plot(kind='pie',
                            figsize=(15,6),
                            autopct="%.1f%%",
                            startangle=90,
                            shadow=True,
                            labels=None,
                            pctdistance=1.15,
                            explode=explode_list);

plt.legend(labels=df_continents.index);
plt.title("Data Imigrasi sesuai Continent", y=1.12);
plt.axis('equal');
```
