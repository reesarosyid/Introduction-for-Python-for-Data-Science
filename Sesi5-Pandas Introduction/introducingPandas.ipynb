### Agregasi

```python
kondisi = df['team_id'] == "MNL"
df[kondisi]['date_game'].agg(("min","max"))
```

## Perbedaan loc dan iloc

```python
data = [0, 0.25, 0.5, 0.75, 1]
ser_data_idx =pd.Series(data, index=['a','b','c','d','e'])
ser_data_idx
```

### Loc

Memanggil data dengan explisit index

```python
ser_data_idx.loc['c']
```

```python
# Slicing loc 
ser_data_idx.loc['c':'e']
```

### Iloc

Memanggil data dengan implisit index

```python
ser_data_idx.iloc[1]
```

```python
# Slicing Iloc
ser_data_idx.iloc[1:4]
```
