# Probabilitas Distribusi

Distribusi probabilitas adalah sebuah susunan distribusi yang mempermudah mengetahui probabilitas sebuah peristiwa. Merupakan hasil dari setiap peluang peristiwa. Sebuah ukuran atau besaran yang merupakan hasil suatu percobaan atau kejadian yang terjadi acak atau untung- untungan dan mempunyai nilai yang berbeda-beda.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from IPython.display import Math, Latex
from IPython.core.display import Image
import seaborn as sns
sns.set(color_codes=True)
sns.set(rc={'figure.figsize':(5,5)})
```

## Uniform Distribusi

memodelkan kejadian yang memiliki kemungkinan sama untuk seluruh kejadian yang ada. Sebagai contoh adalah kejadian melempar dadu.

### Gambar Kurva

```python
from scipy.stats import uniform
```

```python
#Angka acak dari uniform distribusion

n = 10000
start = 10
width = 20
data_uniform = uniform.rvs(size=n, loc=start, scale=width)
```

```python
ax = sns.distplot(data_uniform,
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth":15, 'alpha':1})
ax.set(xlabel = 'Uniform Distribution', ylabel = 'Frequency')
```

## Distribusi Normal

Distribusi normal memiliki kurva kerapatan berbentuk lonceng yang dijelaskan dengan mean dan standar deviasi. Kurva kepadatan simetris, berpusat di sekitar meannya, dengan penyebarannya ditentukan oleh devisasi standarnya yang menunjukna bahwa data di dekat mean lebih sering terjadi dari pada data yang jauh dari mean.

### Gambar Kurva

```python
from scipy.stats import norm
```

```python
data_normal = norm.rvs(size=10000, loc=0, scale=1)
```

```python
ax = sns.distplot(data_normal,
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth":15, 'alpha':1})
ax.set(xlabel = 'Normal Distribution', ylabel = 'Frequency')
```

## Distribusi Poisson

Poisson distribution memodelkan variasi kemungkinan suatu total kejadian pada suatu rentang waktu. Sebagai contoh Jika kita mengetahui kejadian “Banjir” terjadi, secara rata-rata, “3” kali per “bulan”. Maka dapat kita modelkan variasi kemungkinan terjadinya “banjir” dari rata-rata kejadian yang terjadi dan dapat juga melihat batas minimum ataupun maksimum total kejadian yang dapat terjadi pada suatu rentang waktu.

### Gambar kurva

```python
from scipy.stats import poisson
```

```python
data_poisson = poisson.rvs(mu=3, size=10000)
```

```python
ax = sns.distplot(data_poisson,
                  bins=100,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth":15, 'alpha':1})
ax.set(xlabel = 'Poisson Distribution', ylabel = 'Frequency')
```

# Confidence Interval

Confidence Interval adalah sesuatu parameter yang digunakan untuk menentukan keakuratan mean suatu sample. Ketika kita melakukan sejumlah pengukuran pada sebuah sample dan menghitung nilai rata-rata pengukuran tersebut, kita dapat memperkirakan nilai aktual untuk pengukuran tersebut.

$$CI =\bar{x} \mp z_{\alpha/2} \frac{s}{\sqrt{n}}$$

## CI proposisi populasi di Python

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('heart.csv')
df
```

```python
# Mengubah value 1 menajdi Male dan 0 menjadi Female
df['Sex1'] = df.sex.replace({1: "Male", 0: "Female"})
df['Sex1']
```

```python
# Drop nan value
dx = df[["target", "Sex1"]].dropna()
```

```python
# Melakukan cross table target menjadi index dan Sex1 menjadi feature
pd.crosstab(dx.target, dx.Sex1)
```

Disini terlihat bahwa ada 226 wanita yang terkena penyakit jantung. Oleh karena itu kita akan menghitung interval kepercayaan dari proposisi wanita yang terkena penyakit jantung.

```python
# Menghitung proposisi wanita
p_fm = 226/(86+226)
p_fm
```

```python
# Menghitung total populasi
n = 86+226
n
```

```python
# menghitung standar error
se_female = np.sqrt(p_fm*(1-p_fm)/n)
se_female
```

```python
# menghitung interval kepercayaan

z_score = 1.96 # menggunakan z score 96%
lcb = p_fm-z_score*se_female #limit CI terbawah
ucb = p_fm+z_score*se_female #limit CI teratas
lcb, ucb
```

Berarti dapat kita simpulkan interval kepercayaan dari populasi wanita yang terkena penyakit jantung adalah dari 0,67 sampai 0,77.

```python
# Menggunakan statsmodels

import statsmodels.api as sm
sm.stats.proportion_confint(n*p_fm, n)
# akan mendapatkan hasil yang sama
```

## CI rata rata populasi

Kita akan menghitung interval kepercayaan dari rata-rata kolesterol wanita.

```python
df.groupby("Sex1").agg({"chol": [np.mean, np.std, np.size]})
```

```python
mean_fe = 261.45
sd = 64.4
n = 312
z = 1.96
```

```python
# Menghitung standar eror rata rata
se = sd/np.sqrt(n)
se
```

```python
# menghitung interval kepercayaan
lcb = mean_fe-z*se #limit CI terbawah
ucb = mean_fe+z*se #limit CI teratas
lcb, ucb
```

Berarti dapat disimpulkan bahwa rata-rata kolesterol dari populasi perempuan akan jatuh antara 254.303 dan 268.596.

# Hipotesis Testing

Uji hipotesis adalah sebuah proses untuk melakukan evaluasi kekuatan bukti dari sampel, dan memberikan dasar untuk membuat keputusan terkait dengan populasinya. Tujuan uji hipotesis adalah untuk memutuskan apakah hipotesis yang diuji ditolak atau diterima.

## Contoh Hipotesis testing pada python

```python
da = pd.read_csv('nhanes_2015_2016.csv')
da.head()
```

**Research Question**
Dengan mempertimbangkan data NHANES, apakah pria memiliki Body Mass Index rata-rata yang lebih tinggi secara signifikan dari pada wanita?

H0: x'1=x'2

H1: x'1!=x'2

```python
#Membuat kondisi dari data set untuk memisahkan gender pria dan wanita
wanita = da[da["RIAGENDR"]==2]
pria = da[da["RIAGENDR"]==1]
```

```python
# Mengetahui jumlah, rata-rata, dan standar deviasi dari wanita
nWanita = len(wanita)
mWanita = wanita["BMXBMI"].mean()
sdWanita = wanita["BMXBMI"].std()

print(f"Total wanita: {nWanita}")
print(f"Rata-rata wanita: {mWanita}")
print(f"Standar deviasi wanita:{sdWanita}")
```

```python
# Mengetahui jumlah, rata-rata, dan standar deviasi dari pria
nPria = len(pria)
mPria = pria["BMXBMI"].mean()
sdPria = pria["BMXBMI"].std()

print(f"Total pria: {nPria}")
print(f"Rata-rata pria: {mPria}")
print(f"Standar deviasi pria:{sdPria}")
```

```python
# Melakukan hipotesis testing
sm.stats.ztest(wanita["BMXBMI"].dropna(), pria["BMXBMI"].dropna(), alternative="two-sided")
```

**Kesimpulan**
Karena p-value 6.591544431126401e-10 lebih kecil dari alpha 5%, maka kita dapat menolak H0 bahwa rata-rata BMI laki-laki sama dengan perempuan. Dengan catatan, kita menggunakan alternative ="two-sided" dalam z-test karena kita memeriksa ketidaksamaan.
